### Checking for daylight or dark

```python
df_cabi_path['start_date_time'] = pd.to_datetime(df_cabi_path['started_at'])
```

```python
import datetime

def classify_day_or_night(row):
    sunrise_sunset_times = {
        'Jan': ("08:26", "17:57"),
        'Feb': ("08:14", "18:29"),
        'Mar': ("07:40", "19:01"),
        'Apr': ("06:52", "19:32"),
        'May': ("06:10", "20:00"),
        'Jun': ("05:43", "20:37"),
        'Aug': ("06:09", "20:19"),
        'Sep': ("06:37", "19:38"),
        'Oct': ("07:04", "18:50"),
        'Nov': ("07:50", "17:54"),
        'Dec': ("08:23", "17:49"),
    }

    # Extract the month and hour from the 'started_at' column
    started_at = row['start_date_time']
    month_abbrev = started_at.strftime("%b")
    hour = started_at.strftime("%H:%M")

    # Get the sunrise and sunset times for the corresponding month
    sunrise_time, sunset_time = sunrise_sunset_times.get(month_abbrev, ("06:00", "18:00"))

    # Check if the trip started during the night
    if hour < sunrise_time or hour > sunset_time:
        return "Night"
    else:
        return "Day"
```

```python
# Create the 'day_or_night' column
df_cabi_path['day_or_night'] = df_cabi_path.apply(classify_day_or_night, axis=1)
```
